# 📥 STEP 3: Download Dataset

⏱️ **Time: 30 minutes to 2 hours** (depending on internet speed)

**What this does:**
- Downloads 140K high-quality face images from Kaggle
- Organizes them into `real/` and `fake/` folders
- Verifies dataset quality

**Run this cell to download:**

````python
import os
import shutil
from pathlib import Path

print("📥 Dataset Download Assistant")
print("="*60)

# Create download directory
os.makedirs('/tmp/datasets_downloaded', exist_ok=True)

print("\n🎯 Recommended Datasets:")
print("\n1. 140k-real-and-fake-faces (RECOMMENDED) ⭐")
print("   Size: ~50GB (takes 1-2 hours)")
print("   Images: 140,000")
print("   Expected Accuracy: 93-97%")

print("\n2. real-and-fake-face-detection (FAST)")
print("   Size: ~1GB (takes 10-20 mins)")
print("   Images: 7,000")
print("   Expected Accuracy: 75-85%")

print("\n" + "="*60)
print("\nEdit the code below to select which dataset to download.")
print("\n⚠️ WARNING: Don't close this tab while downloading!")
````

````python
# ===== DOWNLOAD DATASET (Choose ONE) =====

print("🔽 Downloading dataset from Kaggle...")
import subprocess
import os

os.chdir('/tmp')

# OPTION A: Recommended (140K images, best accuracy)
dataset_name = '140k-real-and-fake-faces'

# OPTION B: Fast (7K images, good for testing)
# dataset_name = 'real-and-fake-face-detection'

print(f"\n📥 Downloading: {dataset_name}")
print("   This will take a while... ☕")
print("   DO NOT CLOSE THIS TAB!")

result = subprocess.run(
    ['kaggle', 'datasets', 'download', '-d', dataset_name],
    capture_output=False,
    text=True
)

if result.returncode == 0:
    print(f"\n✅ Download complete!")
else:
    print(f"\n❌ Download failed. Check your internet connection.")
````

````python
# ===== EXTRACT AND ORGANIZE DATASET =====

import os
import shutil
from pathlib import Path
from zipfile import ZipFile

print("📦 Extracting and organizing dataset...")

os.chdir('/tmp')

# Find the zip file
zip_files = list(Path('/tmp').glob('*.zip'))

if not zip_files:
    print("❌ No zip file found. Make sure download completed.")
else:
    zip_file = zip_files[0]
    print(f"Found: {zip_file.name}")
    
    # Extract
    print(f"\n📂 Extracting {zip_file.name}...")
    with ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall('/tmp/extracted')
    
    # Create dataset directories
    os.makedirs('/tmp/dataset/real', exist_ok=True)
    os.makedirs('/tmp/dataset/fake', exist_ok=True)
    
    # Move files
    extracted_path = Path('/tmp/extracted')
    
    for item in extracted_path.rglob('*'):
        if item.is_file():
            # Simple logic: files with 'real' in name go to real/, others to fake/
            if 'real' in item.name.lower() or 'authentic' in item.name.lower():
                dest = Path('/tmp/dataset/real') / item.name
            else:
                dest = Path('/tmp/dataset/fake') / item.name
            
            if not dest.exists():
                shutil.copy2(item, dest)
    
    print("\n📊 Dataset Organization Summary:")
    
    real_count = len(list(Path('/tmp/dataset/real').glob('*.*')))
    fake_count = len(list(Path('/tmp/dataset/fake').glob('*.*')))
    total = real_count + fake_count
    
    print(f"   Real images: {real_count}")
    print(f"   Fake images: {fake_count}")
    print(f"   Total: {total}")
    print(f"   Balance: {real_count/total*100:.1f}% real, {fake_count/total*100:.1f}% fake")
    
    # Get folder size
    import subprocess
    size_result = subprocess.run(['du', '-sh', '/tmp/dataset'], capture_output=True, text=True)
    print(f"   Total size: {size_result.stdout.strip()}")
    
    print("\n✅ Dataset ready to use!")
````

# 📊 STEP 5: Check Your Results

After training completes, run this cell to verify your accuracy:

````python
# ===== VERIFY TRAINING RESULTS =====

import json
from pathlib import Path

print("📊 Training Results Summary")
print("="*60)

# Check if metrics file exists
metrics_path = Path('/tmp/model_metrics_full.json')

if metrics_path.exists():
    with open(metrics_path, 'r') as f:
        metrics = json.load(f)
    
    print(f"\n✅ Model Trained Successfully!")
    print(f"\nPerformance Metrics:")
    print(f"  Accuracy:  {metrics.get('accuracy', 'N/A')}")
    print(f"  Precision: {metrics.get('precision', 'N/A')}")
    print(f"  Recall:    {metrics.get('recall', 'N/A')}")
    print(f"  F1-Score:  {metrics.get('f1', 'N/A')}")
    
    if float(str(metrics.get('accuracy', '0')).rstrip('%')) > 85:
        print("\n🎉 EXCELLENT! Your model achieved excellent accuracy!")
    else:
        print("\n⚠️ Accuracy still low. Check dataset size and training epochs.")
else:
    print("\n❌ Metrics file not found.")
    print("   Make sure you ran Section 13 in PROJECT_DOCUMENT.ipynb")

print("\n" + "="*60)
````
